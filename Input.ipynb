```python
import tensorflow as tf
from tensorflow import keras
import IPython.display as ipd
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
import os
```

```python
i=0
a = "Input_to_input/Indian_Grey_Hornbill.mp3"
signal, sr = librosa.load(a, duration=10)  # sr = sampling rate
#print(x)
#plot recording signal
plt.figure(figsize=(10, 4))
#librosa.display.waveplot(signal, sr=sr)
#plt.title("Sound")
#plt.show()
  
# Plot spectogram
plt.figure(figsize=(10, 4))
D = librosa.amplitude_to_db(np.abs(librosa.stft(signal)), ref=np.max)
librosa.display.specshow(D, y_axis="linear")
plt.colorbar(format="%+2.0f dB")
plt.title("Linear-frequency power spectrogram")

# saving the spectrogram
#name = "_India_Grey_Hornbill"
path = "Spectrogram/Class/valid/"
i+=1
letter = str(3)
filename = path+letter
plt.savefig(filename)
plt.show()
```
